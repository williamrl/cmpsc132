We will start with a simple particle class. Feel free to add to this Class as much as you deem fit. 

```python
from dataclasses import dataclass

@dataclass
class Particle:
  '''Class for keeping track of position and mass of a particle'''
  x: float
  y: float
  mass: float
```

The **Particle** class is just the $x,y$ Cartesian coordinates and a dummy variable for mass (this is just used for visulization for now, but can be used for more interesting behaviors).

To initialize the field, we will employ the **Random** library from *Numpy*. 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

random = np.random
random.seed(2022)

#Define the height and width of system
dy, dx = 100,100

numParts = 1000
particles    = ['']*numParts
for i in range(numParts):
  particles[i] = Particle(random.rand()*dy,random.rand()*dx, random.rand())

print(particles[100])
```

Now that we have put a bunch of particles in the field, we can do some simple visulization of the particles. 

```python
fig = plt.figure()
for p in particles:
  plt.scatter(p.x, p.y, s=p.mass*dy, cmap= cm.coolwarm_r)
  
plt.grid('on')
plt.clim(0,1)
fig.set_size_inches(12,12)
plt.show()
```
